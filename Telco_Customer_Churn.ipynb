```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/awithaaa/telco-customer-churn/refs/heads/main/Telco-Customer-Churn.csv")
```

```python
df.isnull().sum()
```

## 1.5 Examine Inconsistencies

```python
for col in df.columns:
  print(col, df[col].unique())
```

- MonthlyCharges are in String, but clearly it should be int.
- If there is 'No' PhoneService then MultipleLines, InternetService, OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingTV AND StreamingMovies could be 'No'

## 1.6 Handling Missing Values
- After converting 'TotalCharges' into numerical values, contains 11 missing values
- The missing values were replaced with the meadian of the TotalCharges

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(df.isna().sum())


df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

### PhoneService vs MutipleLines

```python
combo_counts = df.groupby(['PhoneService', 'MultipleLines']).size().reset_index(name='count')

plt.figure(figsize=(10,6))
sns.barplot(data=combo_counts, x='PhoneService', y='count', hue='MultipleLines')

plt.title("PhoneService vs MultipleLines")
plt.ylabel("Count")
plt.xlabel("PhoneService")
plt.show()
```

### PhoneService vs InternetService

```python
combo_counts = df.groupby(['PhoneService', 'InternetService']).size().reset_index(name='count')

plt.figure(figsize=(10,6))
sns.barplot(data=combo_counts, x='PhoneService', y='count', hue='InternetService')

plt.title("PhoneService vs InternetService")
plt.ylabel("Count")
plt.xlabel("PhoneService")
plt.show()
```

### InternetService vs OnlineSecurity

```python
combo_counts = df.groupby(['InternetService', 'OnlineSecurity']).size().reset_index(name='count')

plt.figure(figsize=(10,6))
sns.barplot(data=combo_counts, x='InternetService', y='count', hue='OnlineSecurity')

plt.title("InternetService vs OnlineSecurity")
plt.ylabel("Count")
plt.xlabel("InternetService")
plt.show()
```

## 1.7 Target Variable Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()

df['Churn'].value_counts(normalize=True)
```

The chrun distribution is imbalanced
- 26.5% churned
- 73.5% not churned

This can be lead to bias toward predicting "No" more often.

## 1.8 Univariate Analysis

### 1.8.1 Numerical Features

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

fig, axes = plt.subplots(1, len(num_cols), figsize=(18, 6))
for k in range(len(num_cols)):
    col = num_cols[k]
    sns.histplot(df[col], bins=30, kde=True, ax=axes[k])
    axes[k].set_title(f'Histogram of {col}')


plt.show()
```

### 1.8.2 Categorical Variables

```python
cat_cols = df.nunique()[df.nunique() < 10].index.tolist()
cat_cols.remove('Churn')

plt.figure(figsize=(15,15))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(5,4,i)
    sns.countplot(y=col, data=df)
    plt.tight_layout()
plt.show()
```

## 1.9 Bivariate Analysis

### 1.9.1 Contract Type vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Contract Type vs Churn")
plt.show()
```

**Insights:**
- Month-to-month contracts have the highest churn rate.
- One-year and two-year contracts strongly reduce churn.

### 1.9.2 Internet Service vs Churn

```python
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Internet Service vs Churn")
plt.show()
```

**Insights:**
- Fiber optic customers churn the most.
- DSL customers churn less.

### 1.9.3 Monthly Charges vs Churn

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

**Insight:**
- Churned customers tend to have higher monthly charges.

### 1.9.4 Tenure vs Churn

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

**Insight:**
- Customers with shorter tenure churn significantly more.

## 1.10 Correlation Analysis

```python
plt.figure(figsize=(6,5))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Numerical Feature Correlation")
plt.show()
```

- Tenure and TotalCharges show strong correlation
- MonthlyCharges and TotalCharges show moderate correlation
